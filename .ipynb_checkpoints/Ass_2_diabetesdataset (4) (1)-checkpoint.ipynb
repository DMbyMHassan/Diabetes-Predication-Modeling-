## 1.b Handle missing values and outliers effectively.

```python
#I will handel zeros for BloodPressure, Glucose, BMI, SkinThickness, however I left Pregnacies and 
#insulin becuase it could be not missng data for diabete type 1 the patient donot have insulin and take insulin injection so it could happen
#Calculate the means for the specified columns

mean_values = df[['BloodPressure', 'Glucose', 'BMI', 'SkinThickness']].replace(0, np.nan).mean()

# Replace zero values with the means
for column in ['BloodPressure', 'Glucose', 'BMI', 'SkinThickness']:
    df[column] = df[column].replace(0, mean_values[column])


df.head()
```
